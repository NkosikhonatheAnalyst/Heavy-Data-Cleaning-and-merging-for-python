```python
#DATA FRAME ONE WHO IS THE FIRST WORKSHEET, THE FIRST THING WE DOING IS TO REMOVE THE EMPTY COLUMNS.
```

```python
df2.drop(columns= ['Unnamed: 0','Unnamed: 1','Unnamed: 6','Unnamed: 7'], inplace=True)
df2
```

```python
#WE WANT TO SEE ALL THE COLUMNS WE HAVE ON DATA FRAME 2.
```

```python
df2.columns
```
